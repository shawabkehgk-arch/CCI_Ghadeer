### Part 1 -- PatientIntake Class (30%)

Define a `PatientIntake` class with:
- **Attributes:** name, mrn, age, gender, diagnosis, hemoglobin, wbc, platelets, creatinine, medications (list of strings), clinical_note (string)
- **Methods:**
  - `get_lab_alerts()` -- returns a list of alert strings for abnormal lab values
  - `is_high_risk()` -- returns True if multiple criteria are met (e.g., low HGB + abnormal WBC + on chemo)
  - `summary()` -- returns a formatted string with all patient info
  - `to_dict()` -- returns a dictionary for CSV export

```python
import pandas as pd

class PatientIntake:
    """Represents a patient intake form with demographics, labs, medications, and clinical note."""

    def __init__(self, name, mrn, age, gender, diagnosis, hemoglobin, wbc,
                 platelets, creatinine, medications=None, clinical_note=""):
        self.name = name
        self.mrn = mrn
        self.age = age
        self.gender = gender
        self.diagnosis = diagnosis
        self.hemoglobin = hemoglobin
        self.wbc = wbc
        self.platelets = platelets
        self.creatinine = creatinine
        self.medications = medications if medications is not None else []
        self.clinical_note = clinical_note

    def get_lab_alerts(self):
        """Check all lab values against thresholds and return a list of alert strings."""
        alerts = []
        # TODO: Implement lab alert logic using CONFIGURATION thresholds
        # For now, a placeholder to avoid errors
        if self.hemoglobin < 10: # Example threshold
            alerts.append("Low Hemoglobin")
        return alerts

    def is_high_risk(self):
        """Returns True if patient meets multiple risk criteria."""
        # TODO: Implement high-risk criteria
        # For now, a placeholder
        return len(self.get_lab_alerts()) > 0 # Example: any alert means high risk

    def summary(self):
        """Returns a formatted summary string."""
        meds_str = ", ".join(self.medications) if self.medications else "None"
        return (f"Patient Name: {self.name} (MRN: {self.mrn})\n" +
                f"Age: {self.age}, Gender: {self.gender}, Diagnosis: {self.diagnosis}\n" +
                f"Labs: HGB={self.hemoglobin}, WBC={self.wbc}, Platelets={self.platelets}, Creatinine={self.creatinine}\n" +
                f"Medications: {meds_str}\n" +
                f"Clinical Note: {self.clinical_note}")

    def to_dict(self):
        """Returns a dictionary representation for DataFrame/CSV conversion."""
        return {
            'name': self.name,
            'mrn': self.mrn,
            'age': self.age,
            'gender': self.gender,
            'diagnosis': self.diagnosis,
            'hemoglobin': self.hemoglobin,
            'wbc': self.wbc,
            'platelets': self.platelets,
            'creatinine': self.creatinine,
            'medications': ", ".join(self.medications),
            'clinical_note': self.clinical_note,
            'is_high_risk': self.is_high_risk(),
            'alert_count': len(self.get_lab_alerts())
        }
```

#### Step C: Check for Drug Interactions

```python
# Step C: Match patient medications against interaction table
# TODO: For each patient, check if any of their medications appear in
# the drug interaction table (as drug1 or drug2)
# Build a list of interaction alerts

drug_interaction_alerts = []

for patient in patients:
    patient_meds = set(patient.medications) # Use a set for efficient lookup

    # Check for interactions where patient's medication is drug1
    for med1 in patient_meds:
        possible_interactions = interactions[interactions['drug1'] == med1]
        for _, row in possible_interactions.iterrows():
            med2_interacting = row['drug2']
            if med2_interacting in patient_meds:
                # Found an interaction
                drug_interaction_alerts.append({
                    'mrn': patient.mrn,
                    'patient_name': patient.name,
                    'drug1': med1,
                    'drug2': med2_interacting,
                    'interaction_severity': row['interaction_severity'],
                    'description': row['description']
                })

    # Check for interactions where patient's medication is drug2 (and not already caught as drug1)
    # This handles symmetric interactions or cases where the patient's drug was 'drug2' in the table
    for med2 in patient_meds:
        possible_interactions = interactions[interactions['drug2'] == med2]
        for _, row in possible_interactions.iterrows():
            med1_interacting = row['drug1']
            if med1_interacting in patient_meds and not any(d['drug1'] == med1_interacting and d['drug2'] == med2 for d in drug_interaction_alerts):
                # Found an interaction not already recorded (to avoid duplicates for symmetric interactions)
                drug_interaction_alerts.append({
                    'mrn': patient.mrn,
                    'patient_name': patient.name,
                    'drug1': med1_interacting,
                    'drug2': med2,
                    'interaction_severity': row['interaction_severity'],
                    'description': row['description']
                })

# Convert the list of alerts to a DataFrame
df_drug_interactions = pd.DataFrame(drug_interaction_alerts)

print(f"Found {len(df_drug_interactions)} potential drug interactions.")
if not df_drug_interactions.empty:
    print(df_drug_interactions.head())
else:
    print("No drug interactions found.")
```

## Part 3 -- Analysis & Critical Thinking (20%)




1. **Failure modes:** The current lab alert logic exhibits two primary failure modes related to diagnostic accuracy:Reduced Specificity (Over-flagging): The logic relies on fixed thresholds that do not account for a patient’s clinical baseline. For example, a patient with a thalassemia trait or long-standing chronic anemia may be erroneously flagged as "high-risk" despite being clinically stable.LLM Solution: This can be mitigated using Few-Shot Prompting, providing the model with examples of chronic vs. acute states to ensure it only alerts on significant deviations from the patient's historical baseline.Reduced Sensitivity (Under-classification): The system utilizes total white blood cell (WBC) counts rather than the Absolute Neutrophil Count (ANC). This creates a diagnostic blind spot where a patient with a normal total WBC but severe neutropenia would be missed.LLM Solution: Chain-of-Thought (CoT) reasoning can be employed to instruct the model to "think step-by-step," calculating the ANC from the laboratory differential before determining the risk level, thereby catching cases the simple threshold logic ignores.

2. **Missing checks:** The notebook currently omits several critical parameters required for comprehensive oncology triage:

Laboratory & Physiological Values: It does not evaluate electrolytes (potassium, sodium), liver function tests (LFTs), or vital signs such as temperature and performance status.

Clinical Context: The system lacks data on patient allergies, current treatment line, and recent hospitalization history.

Unstructured Data: The logic fails to synthesize the free-text clinical notes. In real-world practice, symptoms documented in prose—such as fever, active hemorrhaging, or chest pain—often carry higher clinical urgency than a laboratory pattern alone.

3. **PHI/Security:** Deployment at an institution like the King Hussein Cancer Center (KHCC) necessitates a rigorous security posture to protect Protected Health Information (PHI).

Data Integrity: Identifiable data such as patient names, Medical Record Numbers (MRNs), diagnoses, and specific medication regimens must be handled within a secure, audited environment. This includes implementing Role-Based Access Control (RBAC), data encryption at rest and in transit, and ensuring no public sharing of the notebook.

API Key Management: If an LLM is utilized for clinical note summarization, the API key must never be hard-coded within the notebook. Instead, it should be stored in a secure secret manager or accessed via environment variables to prevent credential exposure.

4. **Testing:** To ensure the notebook performs reliably on new or unseen data, a multi-tiered validation framework should be implemented:

Unit Testing: Develop specific tests for the get_lab_alerts(), is_high_risk(), and drug-interaction functions using known clinical edge cases.

Data Validation: Create automated checks to validate CSV outputs and ensure the system gracefully handles missing, malformed, or null values.

Regression Testing: Systematic verification must be conducted whenever laboratory thresholds or logic parameters are modified to ensure that updates do not introduce new diagnostic errors.
